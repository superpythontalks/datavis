## A short interlude into how to make interactive plots

```python
%matplotlib notebook

import numpy as np
import matplotlib.pyplot as plt
plt.style.use('v_paper')
```

```python
from ipywidgets import interact
```

```python
fig, ax = plt.subplots(2, 1, sharex=True)
x = np.linspace(0,1,100)

@interact
def interactive_plot(f:(0.1,10)):
    y0 = np.sin(x*f)
    y1 = np.cos(x*f)
    ax[0].clear()
    ax[0].plot(x, y0)
    ax[1].clear()
    ax[1].plot(x, y1)
```
